```python
import numpy as np
import matplotlib.pyplot as plt

#plotting
plt.rcParams['text.usetex'] = True
plt.rcParams.update({'font.size': 22})
```

Bifurcation diagram for Question 1

```python
muc = 1/3
mus = np.arange(-1,1,0.01)
mus_pos = np.arange(-1,1,0.01)
```

```python
fig = plt.figure(figsize=(8,8))

plt.hlines(0, -1, 1, color='r')
plt.plot(np.arange(-1,muc,0.01), np.arange(-1,muc,0.01), 'r--')
plt.plot(np.arange(muc,1,0.01), np.arange(muc,1,0.01), 'r')
plt.plot(mus_pos, -np.sqrt(mus_pos/3), 'r--')
plt.plot(np.arange(0,muc,0.01), np.sqrt(np.arange(0,muc,0.01)/3), 'r')
plt.plot(np.arange(muc,1,0.01), np.sqrt(np.arange(muc,1,0.01)/3), 'r--')
plt.scatter(0,0,color='k',marker='s',s=80)
plt.scatter(muc,muc,color='k',marker='s',s=80)
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.xlabel('$\mu$')
plt.ylabel('$x^\star$',rotation=0)
plt.title('Bifurcation diagram for $\dot{x} = x(x-\mu)(\mu-3 x^2)$',fontsize=21)
plt.grid()
```
